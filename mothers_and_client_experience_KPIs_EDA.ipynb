```python
# Import necessary libraries

import pandas as pd
```

```python
# Load the csv file

prompts_clients_df = pd.read_csv("data/prompts_clients.csv")
prompts_clients_df.head()
```

#### A. Core KPIs

##### 1. Reach and engagement

```python
# 1. Average number of clients per month

prompts_clients_df["interaction_date"] = pd.to_datetime(prompts_clients_df["interaction_date"], errors="coerce"
)

# Year–month
prompts_clients_df["interaction_year_month"] = prompts_clients_df["interaction_date"].dt.to_period("M")

# Active clients per month = distinct client_id per month
active_per_month = prompts_clients_df.groupby("interaction_year_month")["client_id"].nunique()

# Average number of active clients per month
average_active_clients_per_month = active_per_month.mean()

print(average_active_clients_per_month)
```

###### In the eight counties where the program operates, an average of 173 clients use PROMPTS each month to access maternal health services, indicating meaningful engagement but also highlighting room to scale outreach and enrollment further.
